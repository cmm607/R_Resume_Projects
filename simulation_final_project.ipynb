```python


grand_canyon_data = master_data[master_data['BU_simplified_name'] == 'GRAND CANYON']

# Print the filtered data
grand_canyon_data.head()
```

```python
#the average of the margin_rate column in grand_canyon_data

print(grand_canyon_data['margin_rate'].mean())
```

```python
#calc total cost
grand_canyon_data['total_cost'] = grand_canyon_data['units_sold'] * grand_canyon_data['net_single_unit_cost']
```

```python
#aggregate total cost by week
weekly_total_cost = grand_canyon_data.groupby('fiscal_week_end_date')['total_cost'].sum()
```

```python
#check normality test for total_cost
import scipy.stats as stats
stat, p = stats.shapiro(weekly_total_cost)

# Interpret the results
alpha = 0.05
if p > alpha:
    print('The data appears to be normally distributed (fail to reject H0).')
else:
    print('The data does not appear to be normally distributed (reject H0).')
```

```python
#minimum, mean, and maximum of the unit_price column of grand_canyon_data

print(f"Minimum unit price: {grand_canyon_data['unit_price'].min()}")
print(f"Mean unit price: {grand_canyon_data['unit_price'].mean()}")
print(f"Maximum unit price: {grand_canyon_data['unit_price'].max()}")
```

```python

# aggregate units_sold weekly

weekly_units_sold = grand_canyon_data.groupby('fiscal_week_end_date')['units_sold'].sum()
print(weekly_units_sold)
```

```python
# aggregate gross_margin weekly

weekly_margin = grand_canyon_data.groupby('fiscal_week_end_date')['gross_margin'].sum()

print(weekly_margin)
```

```python
#avg weekly price

weekly_unit_price = grand_canyon_data.groupby('fiscal_week_end_date')['unit_price'].mean()
print(weekly_unit_price)
```

```python
#weekly unit cost

weekly_unit_cost = grand_canyon_data.groupby('fiscal_week_end_date')['net_single_unit_cost'].mean()
print(weekly_unit_cost)
```

```python
#merged weekly tables

weekly_table = pd.concat([weekly_unit_price, weekly_units_sold, weekly_margin, weekly_unit_cost, weekly_total_cost ], axis=1).reset_index()

weekly_table.head()
```

```python
# Calculate the percent change in weekly_units_sold and weekly_unit_price
weekly_table['weekly_units_sold_pct_change'] = weekly_table['units_sold'].pct_change()
weekly_table['weekly_unit_price_pct_change'] = weekly_table['unit_price'].pct_change()

# Calculate the price elasticity of demand and adding it in as a column
weekly_table['price_elasticity_of_demand'] = weekly_table['weekly_units_sold_pct_change'] / weekly_table['weekly_unit_price_pct_change']
weekly_table.head()
```

```python
#the average of the unit_price column in weekly_table

print(weekly_table['unit_price'].mean())
```

```python
#minimum, mean, and maximum of the unit_price column

print(weekly_table['unit_price'].min())
print(weekly_table['unit_price'].mean())
print(weekly_table['unit_price'].max())
```

```python

#identify the type of distribution of weekly_units_sold - beta is.90 the 2nd hihgest nehind genextreme

import scipy.stats as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data into a pandas DataFrame
data = weekly_units_sold

# Visual inspection
sns.histplot(data, kde=True)
plt.show()

# Define the distributions to test
dist_names = ["norm", "exponweib", "weibull_max", "weibull_min", "pareto", "genextreme", "gamma", "beta", "lognorm"]

# Fit your data to each distribution
dist_results = []
params = {}
for dist_name in dist_names:
    dist = getattr(st, dist_name)
    param = dist.fit(data)
    params[dist_name] = param
    # Apply the Kolmogorov-Smirnov test
    D, p = st.kstest(data, dist_name, args=param)
    print(f"p value for {dist_name} = {p}")
    dist_results.append((dist_name, p))

# Find the best fitting distribution
best_dist, best_p = max(dist_results, key=lambda item: item[1])
print(f"Best fitting distribution: {best_dist}")
print(f"Best p value: {best_p}")
print(f"Parameters for the best fit: {params[best_dist]}")
```

```python
# Monte Carlo simulation to estimate the optimal price that maximizes profit.

import numpy as np
import matplotlib.pyplot as plt

# Define parameters (you can adjust these based on your data)

mean_demand = 251019  # Mean demand
std_demand = 20654    # Standard deviation of demand
mean_cost = 408204.804118
std_cost =  33677.313093
price_range = np.random.normal(weekly_unit_price.mean(), weekly_unit_price.std(), 100) # Possible price range


# Generate random samples for demand
n_iterations = 1000000
total_demand = np.random.normal(mean_demand, std_demand, n_iterations)
total_cost = np.random.normal(mean_cost, std_cost, n_iterations)

# Calculate profit for each combination of price and demand
profits = []
for p in price_range:
    profit = np.mean(p * total_demand - total_cost)
    profits.append(profit)

# Find optimal price that maximizes average profit
optimal_price = price_range[np.argmax(profits)]

# Optional: Plot profit distribution
plt.hist(profits, bins=50, edgecolor='black', alpha=0.7)
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.title('Profit Distribution')
plt.legend()
plt.show()

print(f"Optimal price for maximum profit: ${optimal_price:.2f}")
```
